```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
from google.colab import drive
from PIL import Image, ImageDraw, ImageFont
from google.colab.patches import cv2_imshow

import matplotlib.pyplot as plt
```

```python
class CNN(nn.Module):
  def __init__(self):
    super(CNN, self).__init__()
    self.conv1 = nn.Conv2d(in_channels=1, out_channels=3, kernel_size=5, stride=1)
    self.conv2 = nn.Conv2d(in_channels=3, out_channels=10, kernel_size=5, stride=1)
    self.fc1 = nn.Linear(10 * 12 * 12, 50)
    self.classifier_1 = nn.Linear(50, 10)
    self.classifier_2 = nn.Linear(50, 2)
    self.classifier_3 = nn.Linear(50, 5)

  def forward(self, x):
    x = F.relu(self.conv1(x))
    x = F.relu(self.conv2(x))
    x = x.view(-1, 10 * 12 * 12)
    x = F.relu(self.fc1(x))
    yhat_1 = self.classifier_1(x)
    yhat_2 = self.classifier_2(x)
    yhat_3 = self.classifier_3(x)
    return yhat_1, yhat_2, yhat_3
```
